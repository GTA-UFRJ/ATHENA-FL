# Import the packages

```python
import numpy as np
import tensorflow as tf
from pickle import load
from matplotlib import pyplot
import matplotlib.pyplot as plt

from load_federated_data import *
from generate_neural_network import build_model
```

# Load dataset

```python
dataset_name = "CIFAR-10"
clientID = 1
numClients = 20
basicNN = 0
modelType = 0
trPer = 0.8

x_train, y_train, x_test, y_test = load_data_federated_IID(dataset_name, clientID, numClients, basicNN, modelType, trPer)
```

# Data visualization

```python
print("Training data size: ",len(x_train)," Test data size: ",len(x_test))

unique_train, counts_train = np.unique(y_train, return_counts=True)
unique_test, counts_test = np.unique(y_test, return_counts=True)


print("Training labels: ",unique_train," Test labels: ",unique_test)
print(dict(zip(unique_train, counts_train)))
print(dict(zip(unique_test, counts_test)))

image = x_train[3]
plt.rcParams["figure.figsize"] = (1,1)
plt.imshow(image)
plt.show()
```

# Creating model

```python
model = build_model(basicNN,dataset_name)
```

# Fit the model with the preprocessed data

```python
bs = 32

model.fit(x_train, y_train, epochs=400,batch_size=bs,steps_per_epoch=int(len(x_train)//bs))
```

# Evaluate the final model

```python
loss, accuracy = model.evaluate(x_test,  y_test, verbose=2)
```
